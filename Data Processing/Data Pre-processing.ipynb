```python
import os
import csv
import json
```

```python
def json_to_csv(folder, output):
    json_files = [f for f in os.listdir(folder) if f.endswith('.json')]

    with open(output, 'w', newline='', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)
        
        writer.writerow(['ID',
                         'Top',
                         'Jungle',
                         'Mid',
                         'Bot',
                         'Support',
                         'Enemy Top',
                         'Enemy Jungle',
                         'Enemy Mid',
                         'Enemy Bot',
                         'Enemy Support',
                         'Team Gold',
                         'Team Kills',
                         'Team Deaths',
                         'Dragons',
                         'Dragon Soul',
                         'Baron',
                         'Rift Herald',
                         'First tower',
                         'First Kill',
                         'Team Total CC',
                         'Victory'])

        for json_file in json_files:
            json_path = os.path.join(folder, json_file)

            with open(json_path, 'r', encoding='utf-8') as file:
                try:
                    data = json.load(file)
                    gold_spent = 0
                    CCTime = 0
                    for x in range(0, 5):
                        gold_spent += data["info"]["participants"][x]["goldSpent"]
                        CCTime += data["info"]["participants"][x]["timeCCingOthers"]

                    row_data = [
                        json_file.replace(".json", ""),
                        data["info"]["participants"][0]["championName"],
                        data["info"]["participants"][1]["championName"],
                        data["info"]["participants"][2]["championName"],
                        data["info"]["participants"][3]["championName"],
                        data["info"]["participants"][4]["championName"],
                        data["info"]["participants"][5]["championName"],
                        data["info"]["participants"][6]["championName"],
                        data["info"]["participants"][7]["championName"],
                        data["info"]["participants"][8]["championName"],
                        data["info"]["participants"][9]["championName"],
                        gold_spent,
                        data["info"]["teams"][0]["objectives"]["champion"]["kills"],
                        data["info"]["teams"][1]["objectives"]["champion"]["kills"],
                        data["info"]["teams"][0]["objectives"]["dragon"]["kills"],
                        bool(data["info"]["teams"][0]["objectives"]["dragon"]["kills"] >= 4),
                        data["info"]["teams"][0]["objectives"]["baron"]["kills"],
                        data["info"]["teams"][0]["objectives"]["riftHerald"]["kills"],
                        data["info"]["teams"][0]["objectives"]["tower"]["first"],
                        data["info"]["teams"][0]["objectives"]["champion"]["first"],
                        CCTime,
                        data["info"]["teams"][0]["win"]
                    ]
                    writer.writerow(row_data)
                    
                    gold_spent = 0
                    CCTime = 0
                    for x in range(5, 10):
                        gold_spent += data["info"]["participants"][x]["goldSpent"]
                        CCTime += data["info"]["participants"][x]["timeCCingOthers"]

                    row_data = [
                        json_file.replace(".json", ""),
                        data["info"]["participants"][5]["championName"],
                        data["info"]["participants"][6]["championName"],
                        data["info"]["participants"][7]["championName"],
                        data["info"]["participants"][8]["championName"],
                        data["info"]["participants"][9]["championName"],
                        data["info"]["participants"][0]["championName"],
                        data["info"]["participants"][1]["championName"],
                        data["info"]["participants"][2]["championName"],
                        data["info"]["participants"][3]["championName"],
                        data["info"]["participants"][4]["championName"],
                        gold_spent,
                        data["info"]["teams"][1]["objectives"]["champion"]["kills"],
                        data["info"]["teams"][0]["objectives"]["champion"]["kills"],
                        data["info"]["teams"][1]["objectives"]["dragon"]["kills"],
                        bool(data["info"]["teams"][1]["objectives"]["dragon"]["kills"] >= 4),
                        data["info"]["teams"][1]["objectives"]["baron"]["kills"],
                        data["info"]["teams"][1]["objectives"]["riftHerald"]["kills"],
                        data["info"]["teams"][1]["objectives"]["tower"]["first"],
                        data["info"]["teams"][1]["objectives"]["champion"]["first"],
                        CCTime,
                        data["info"]["teams"][1]["win"]
                    ]
                    writer.writerow(row_data)
                except json.JSONDecodeError:
                    print(f"Error decoding JSON file: {json_file}")
                except KeyError as e:
                    key_error = str(e)
                    print(f"KeyError: '{key_error}' is missing in the data")

    print(f"CSV saved as {output}")

json_folder = '/Users/andre/Documents/Schoolwork/UC Davis/ECS 171/Project Files/games'
output_csv = 'match_data.csv'

json_to_csv(json_folder, output_csv)
```
